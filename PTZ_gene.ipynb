```python
from rdkit import Chem
from rdkit.Chem import AllChem

smiles = "CC(C)=CCN1CC[C@@]2(C)c3cc(O)ccc3C[C@@H]1[C@@H]2C"
mol = Chem.MolFromSmiles(smiles)
mol = Chem.AddHs(mol)  # add hydrogens
AllChem.EmbedMolecule(mol, randomSeed=0xf00d)  # generate 3D coords
result = AllChem.MMFFOptimizeMolecule(mol)    # optimize

if result == 0:
    print("Optimization converged!")
else:
    print("Optimization failed or incomplete.")
```

```python
mol
```

```python
from rdkit.Chem import SDWriter

writer = SDWriter("optimized_molecule.sdf")
writer.write(mol)
writer.close()
```
